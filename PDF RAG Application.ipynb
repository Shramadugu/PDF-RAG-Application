## Step 2: Import the Library


```python
import PyPDF2
import warnings
warnings.filterwarnings("ignore")
```

## Step 3: Define Function to Extract Text from a PDF


```python
def extract_text_from_pdf(pdf_file_path):
    """Extract text from a PDF file."""
    try:
        with open(pdf_file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            return ' '.join(page.extract_text() for page in reader.pages if page.extract_text())
    except Exception as e:
        print(f"Error reading {pdf_file_path}: {e}")
        return ""
```

## Step 4: Define List of PDF File Paths


```python
# List of PDF file paths
pdf_file_path= [
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-1.pdf",
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-2.pdf",
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-3.pdf",
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-4.pdf",
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-5.pdf",
    r"C:\Users\user\Downloads\Document Cloud\Document Cloud\business plan (1)-part-6.pdf"
]

```

## Step 5: Extract Text from Each PDF


```python
# Extract text from the list of PDFs
extracted_texts = []
for pdf_file in pdf_file_path:
    print(f"Processing: {pdf_file}")  # Debugging line
    extracted_text = extract_text_from_pdf(pdf_file)
    extracted_texts.append(extracted_text)

# Print extracted texts
for i, text in enumerate(extracted_texts):
    print(f"--- Text from PDF {i + 1} ---")
    print(text)
    print("\n")
```
